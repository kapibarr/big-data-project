```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Статистика по станциям
## Года
Датасеты по станциям доступны только в 2013-2017 годах.
## Желаемые результаты
- Средняя вместимость станции
- Количество станций
- Вместимость всех станций

```python
capacity_list = []
amount_of_stations = []
for year in range(2013, 2018):
    df = pd.read_csv(f"data/{year}/Divvy_Stations_{year}.csv")
    capacity_list.append(df["dpcapacity"].sum())
    amount_of_stations.append(df.shape[0])
```

```python
plt.plot(range(2013, 2018), capacity_list)
plt.xticks(range(2013, 2018))
plt.title("Вместимость станций полученная по датасетам станций")
plt.xlabel("Год")
plt.ylabel("Вместимость")
```

```python
plt.plot(range(2013, 2018), amount_of_stations)
plt.xticks(range(2013, 2018))
plt.title("Количество станций по датасетам станций")
plt.xlabel("Год")
plt.ylabel("Количество станций")
```

```python
average_capacity_per_station = [
    round(capacity_list[i] / amount_of_stations[i], 2)
    for i in range(len(capacity_list))
]
```

```python
plt.plot(range(2013, 2018), average_capacity_per_station)
plt.xticks(range(2013, 2018))
plt.title("Средняя вместимость станции по годам")
plt.xlabel("Год")
plt.ylabel("Средняя вместимость станции")
```

## Косвенный подсчет станций
Несмотря на то, что нет прямых датасетов с данными по станциям, можно посмотреть на количество уникальных названий станций в датасетах с поездками.

```python
df = pd.read_csv("data/2018/Divvy_Trips_2018_full.csv.gz", compression="gzip")
```

```python
amount_of_stations.append(df["03 - Rental Start Station Name"].unique().shape[0])
```

```python
df = pd.read_csv("data/2019/Divvy_Trips_2019_full.csv.gz", compression="gzip")
```

```python
amount_of_stations.append(df["to_station_name"].unique().shape[0])
```

```python
# Дальше данные не отличаются по структуре
for year in range(2020, 2024):
    df = pd.read_csv(f"data/{year}/Divvy_Trips_{year}_full.csv.gz", compression="gzip")
    amount_of_stations.append(df["end_station_name"].unique().shape[0])
```

```python
plt.plot(range(2013, 2024), amount_of_stations)
plt.title("Количество станций по всем годам подсчитанное косвенно")
plt.xticks(range(2013, 2024))
plt.xlabel("Год")
plt.ylabel("Количество станций")
```
