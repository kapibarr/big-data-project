```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
amount_of_rides = []
```

```python
for year in range(2013, 2024):
    # Берем размер файла каждого года с 2013 до 2023
    if year == 2013:
        df = pd.read_csv("data/2013/Divvy_Trips_2013.csv")
    else:
        df = pd.read_csv(
            f"data/{year}/Divvy_Trips_{year}_full.csv.gz", compression="gzip"
        )
    amount_of_rides.append((year, df.shape[0]))
```

```python
amount_of_rides = sorted(amount_of_rides, key=lambda x: x[0])
```

```python
amount_of_rides
```

```python
# Делим на тысячу, чтобы получить более приятные числа для графика
amount_of_rides = [x[1] / 1000 for x in amount_of_rides]
```

```python
plt.plot(range(2013, 2024), amount_of_rides)
plt.xlabel("Год")
plt.ylabel("Заработок (в тысячах)")
plt.xticks(range(2013, 2024))
plt.title("Количество поездок по годам")
```

Для более точного графика уберем 2023, так как за этот год полной информации нет

```python
plt.plot(range(2013, 2023), amount_of_rides[:-1])
plt.xlabel("Год")
plt.ylabel("Заработок (в тысячах)")
plt.xticks(range(2013, 2023))
plt.title("Количество поездок по годам")
```
